```python
import turtle
import time
import random 
import os

speed = 0.1


#SCORES
score = 0
high_score = 0

#WINDOW SETUP
wn = turtle.Screen()
wn.title("Snake by Edu Vazquez")
wn.bgcolor("black")
wn.setup(width=500, height=500)
wn.tracer(0)

#SNAKE HEAD
head = turtle.Turtle()
head.speed(0) #make appears the head when load the game
head.shape("square")
head.color("white")
head.penup() #erase line of pen
head.goto(0,0)
head.direction = "right"


#FUNCTIONS
#=========================================

#RESET SCORE
def reset_score():
    score = 0
    text.clear()
    text.write("Score: {}       High Score: {}".format(score, high_score), align="center", font=("Courier", 24,"normal"))
    

#KEYBOARD functions
def kb_up():
    head.direction = "up"
def kb_down():
    head.direction = "down"
def kb_right():
    head.direction = "right"
def kb_left():
    head.direction = "left"   

#SNAKE MOVEMENT
def mov():
    if head.direction == "up":
        y = head.ycor()
        head.sety(y + 20)
    if head.direction == "down":
        y = head.ycor()
        head.sety(y - 20)
    if head.direction == "right":
        x = head.xcor()
        head.setx(x + 20)
    if head.direction == "left":
        x = head.xcor()
        head.setx(x - 20)

#FOOD
food = turtle.Turtle()
food.speed(0) #make appears the head when load the game
food.shape("circle")
food.color("white")
food.penup() #erase line of pen
food.goto(0,100)
food.direction = "right"

#BODY GROWT
growt = []

#TEXT
text = turtle.Turtle()
text.speed(0)
text.color("white")
text.penup()
text.hideturtle()
text.goto(0, 220)
text.write("Score: 0        High Score: 0", align="center", font=("Courier", 24,"normal"))

#KEYBOARD CONECTION
wn.listen()
wn.onkeypress(kb_up, "Up")
wn.onkeypress(kb_down, "Down")
wn.onkeypress(kb_right, "Right")
wn.onkeypress(kb_left, "Left")

#=========================================

#MAIN BUCLE (all games has it)
while True:
    wn.update()

    #BORDERS
    if head.xcor() > 250 or head.ycor() > 250 or head.xcor() < -250 or head.ycor() < -250:
        os.system("afplay Basso.aiff&")
        time.sleep(1)
        head.goto(0,0)
        head.direction = "stop"

        #ERASE BODY
        for growt_new in growt:
            growt_new.goto(1000,1000)
        growt.clear()
        
        #RESET SCORE
        reset_score()

    if head.distance(food) < 20:
        y = random.randint(-230,230 )
        x = random.randint(-230,230 )
        food.goto(x,y)

        growt_new = turtle.Turtle()
        growt_new.speed(0) 
        growt_new.shape("square")
        growt_new.color("grey")
        growt_new.penup() #erase line of pen
        growt.append(growt_new)

        #SCORE ADDING
        score += 1
        os.system("afplay Pop.aiff&")

        if score > high_score:
            high_score += 1
        text.clear()
        text.write("Score: {}       High Score: {}".format(score, high_score), align="center", font=("Courier", 24,"normal"))


    #move snake body
    total_growt = len(growt)
    for i in range(total_growt -1, 0, -1):
        x = growt[i -1].xcor()
        y = growt[i -1].ycor()
        growt[i].goto(x,y)
    if total_growt > 0:
        x = head.xcor()
        y = head.ycor()
        growt[0].goto(x,y)
    
    mov()

    #snake collition with body
    for growt_new in growt:
        if growt_new.distance(head) < 20:
            os.system("afplay Basso.aiff&")
            time.sleep(1)
            head.goto(0,0)
            head.direction = "stop"
            reset_score()

            #HIDE BODY
            for growt_new in growt:
                growt_new.goto(10000,1000)
            growt.clear()


    
    time.sleep(speed)
     


```
